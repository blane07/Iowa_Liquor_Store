Once you've chosen your scenario, download the data from the Iowa website in csv format. Start by loading the data with pandas. You may need to parse the date columns appropriately.

```python
import numpy as np
import pandas as pd
from beakerx import *

df = pd.read_csv('~/blane07/GA/iowa_liquor_sales_proj_2.csv', infer_datetime_format=True, dayfirst=True);
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum().sum()
```

```python
df.columns
```

```python
df.drop(['Invoice/Item Number', 'Store Number', 'Address', 'City', 'Zip Code', 
         'Store Location', 'County Number', 'Category', 'Category Name', 'Vendor Name',
         'Item Number', 'Item Description', 'Pack', 'Bottle Volume (ml)', 'State Bottle Cost', 
         'State Bottle Retail', 'Volume Sold (Gallons)'], axis='columns', inplace=True);
```

```python
df = df.sort_values('Store Name').sort_values('County');
```

```python
df.Date = df.Date.apply(lambda x: int(x.split('/')[2] + x.split('/')[0] + x.split('/')[1]));   # YYYYMMDD['Invoice/Item Number', 'Date', 'Store Number', 'Store Name', 'Address',
```

```python
df['Sale (Dollars)'] = df['Sale (Dollars)'].apply(lambda x: float(x.strip('$')));    # Convert string sales to floats
```

```python
df.loc[:, 'Price per Bottle'] = df['Sale (Dollars)'] / df['Bottles Sold']
```

```python
df.loc[:, 'Month'] = df.Date.apply(lambda x: int(str(x)[4:6]))
```

```python
df.loc[:, 'Year'] = df.Date.apply(lambda x: int(str(x)[0:4]))
```

```python
df.drop('Date', axis='columns', inplace=True)
```

```python
df.groupby('Year')['Month'].unique()
```

```python
# Reference: https://www.kdnuggets.com/2017/02/removing-outliers-standard-deviation-python.html

for col in ['Bottles Sold', 'Sale (Dollars)', 'Volume Sold (Liters)', 'Price per Bottle']:    # Filtering outliers
    df = df[(df[col] < np.mean(df[col]) + 3 * np.std(df[col])) & (df[col] > np.mean(df[col]) - 3 * np.std(df[col]))]
```

```python
store_dict_dict = {}
for store in df:
    store_dict = {}
    store_dict['Vendors'] = list(df.groupby('Store Name').get_group(store)['Vendor Number'].unique())
    for year in [2015, 2016]:
        store_dict[(str(year) + ' Sales')] = round(df[df.Year == year].groupby('Store Name').get_group(store)['Sale (Dollars)'].sum(), 2)    
    for year in [2015, 2016]:
        if year == 2015:
            for month in range(1, 13):
                store_dict[(str(year) + str(month))] = round(df[(df.Year == year) & (df.Month == month)].groupby('Store Name').get_group(store)['Sale (Dollars)'].sum(), 2)
        if year == 2016:
            for month in range(1, 4):
                store_dict[(str(year) + str(month))] = round(df[(df.Year == year) & (df.Month == month)].groupby('Store Name').get_group(store)['Sale (Dollars)'].sum(), 2)
    store_dict['Bottles Sold 2015'] = df[df.Year == 2015].groupby('Store Name').get_group(store)['Bottles Sold'].sum()
    store_dict['Bottles Sold 2016'] = df[df.Year == 2016].groupby('Store Name').get_group(store)['Bottles Sold'].sum()
    store_dict['Volume Sold (L) 2015'] = df[df.Year == 2015].groupby('Store Name').get_group(store)['Volume Sold (Liters)'].sum()
    store_dict['Volume Sold (L) 2016'] = df[df.Year == 2016].groupby('Store Name').get_group(store)['Volume Sold (Liters)'].sum()
    store_dict['Avg Price per Bottle 2015'] = round(df[df.Year == 2015].groupby('Store Name').get_group(store)['Price per Bottle'].mean(), 3)
    store_dict['Avg Price per Bottle 2016'] = round(df[df.Year == 2016].groupby('Store Name').get_group(store)['Price per Bottle'].mean(), 3)
    store_dict['Vendors'] = list(df.groupby('Store Name').get_group(store)['County'].unique())
    
    print(pd.DataFrame(store_dict))
```
